```python
from transformers import pipeline
```

```python
mask_filler = pipeline(
    "fill-mask",
    model="bert-base-uncased"
)
```

```python
def fix_sentence(sentence):
    results = mask_filler(sentence)

    print("Suggestions:")
    for res in results[:5]:
        print(f"- {res['token_str']}  (score: {res['score']:.2f})")
```

```python
sentence = "I am [MASK] to school"
fix_sentence(sentence)
```

```python
fix_sentence("She is [MASK] than me")
```

```python
fix_sentence("He has [MASK] his homework")
```
